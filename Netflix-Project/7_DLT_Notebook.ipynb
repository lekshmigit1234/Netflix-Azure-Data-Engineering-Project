#**Create Expectation by adding some rules (Data QUALIOTY CHECKS)**

```python
looktables_rules = {
    "rule1": "show_id IS NOT NULL"
}
```

```python
@dlt.table(
    name = "gold_netflixdirectors"
)
@dlt.expect_all_or_drop(looktables_rules)
def myfunc():
    df = spark.readStream.format("delta").load("abfss://silver@netflixazuredatalake.dfs.core.windows.net/netflix_directors")
    return df
```
