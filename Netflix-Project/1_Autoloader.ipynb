##**Incremental Data Loading Using Autoloader**

```python
%sql
CREATE SCHEMA IF NOT EXISTS netflix_catalog.net_schema
```

```python
checkpoint_location= "abfss://silver@netflixazuredatalake.dfs.core.windows.net/checkpoint"
```

```python
df=spark.readStream\
    .format("cloudFiles")\
    .option("cloudFiles.format","csv")\
    .option("cloudFiles.schemaLocation",checkpoint_location)\
    .load("abfss://raw@netflixazuredatalake.dfs.core.windows.net")

    
```

```python
display(df, checkpointLocation=checkpoint_location)
```

```python
df.writeStream\
    .format("delta") \
    .option("checkpointLocation",checkpoint_location)\
    .trigger(processingTime="10 seconds")\
    .start("abfss://bronze@netflixazuredatalake.dfs.core.windows.net/netflix_titles")
```
